```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
d=d.dropna()
```

```python
d.isnull().sum()
```

```python
d.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
```

```python
d['Loan_Status'].head()
```

```python
d['Dependents'].value_counts()
```

```python
d=d.replace(to_replace='3+',value=4)
```

```python
d['Dependents'].value_counts()
```

```python
d=d.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':1},'Self_Employed':{'No':0,'Yes':1},'Property_Area':{'Rural':0,'SemiUrban':3,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}})
```

```python
d.head()
```

```python
d['Property_Area'].value_counts()
```

```python
d['Property_Area'].replace(to_replace='Semiurban',value=3,inplace=True)
```

```python
d['Property_Area'].value_counts()
```

```python
x=d.drop(columns=['Loan_ID','Loan_Status'],axis=1)
y=d['Loan_Status']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train_prediction=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("accuracy of training data:",training_data_accuracy)
```

```python
x_test_prediction=classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("accuracy on test data:",test_data_accuracy)
```
